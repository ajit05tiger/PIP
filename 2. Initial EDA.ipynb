## 3. Acceleration using change in load

```python
# Force = Mass * Acceleration
# acceleration (a) = Force / Weight of cat (w)

df_ne['acc_lc0']= df_ne['diff_lc0'].apply(lambda x: x/w_ne)
df_ne['acc_lc1']= df_ne['diff_lc1'].apply(lambda x: x/w_ne)
df_ne['acc_lc2']= df_ne['diff_lc2'].apply(lambda x: x/w_ne)
df_ne['acc_lc3']= df_ne['diff_lc3'].apply(lambda x: x/w_ne)
df_ne.head()
```

```python
df_urin['acc_lc0']= df_urin['diff_lc0'].apply(lambda x: x/w_urin)
df_urin['acc_lc1']= df_urin['diff_lc1'].apply(lambda x: x/w_urin)
df_urin['acc_lc2']= df_urin['diff_lc2'].apply(lambda x: x/w_urin)
df_urin['acc_lc3']= df_urin['diff_lc3'].apply(lambda x: x/w_urin)
```

```python
df_def['acc_lc0']= df_def['diff_lc0'].apply(lambda x: x/w_def)
df_def['acc_lc1']= df_def['diff_lc1'].apply(lambda x: x/w_def)
df_def['acc_lc2']= df_def['diff_lc2'].apply(lambda x: x/w_def)
df_def['acc_lc3']= df_def['diff_lc3'].apply(lambda x: x/w_def)
```

```python
X1 = df_ne.timestamp
X2 = df_urin.timestamp
X3 = df_def.timestamp

Y1 = df_ne.acc_lc0
Y2 = df_urin.acc_lc0
Y3 = df_def.acc_lc0

figure, axis = plt.subplots(3,1, figsize=(10,10))

axis[0].plot(X1,Y1)
axis[0].set_title("non extraction")

axis[1].plot(X2, Y2)
axis[1].set_title("urination")

axis[2].plot(X3, Y3)
axis[2].set_title("defecation")

plt.show()
```

## 4. Velocity

```python
df_ne['velo_lc0'] = 0
df_ne['velo_lc1'] = 0
df_ne['velo_lc2'] = 0
df_ne['velo_lc3'] = 0
```

```python
df_ne['velo_lc0'] = 0
df_ne['velo_lc1'] = 0
df_ne['velo_lc2'] = 0
df_ne['velo_lc3'] = 0
for i in range (1,len(df_ne)):
    df_ne['velo_lc0'][i]=(df_ne['acc_lc0'][i]*0.025)+(df_ne['velo_lc0'][i-1])
    df_ne['velo_lc1'][i]=(df_ne['acc_lc1'][i]*0.025)+(df_ne['velo_lc1'][i-1])
    df_ne['velo_lc2'][i]=(df_ne['acc_lc2'][i]*0.025)+(df_ne['velo_lc2'][i-1])
    df_ne['velo_lc3'][i]=(df_ne['acc_lc3'][i]*0.025)+(df_ne['velo_lc3'][i-1])
    
df_urin['velo_lc0'] = 0
df_urin['velo_lc1'] = 0
df_urin['velo_lc2'] = 0
df_urin['velo_lc3'] = 0
for i in range (1,len(df_urin)):
    df_urin['velo_lc0'][i]=(df_urin['acc_lc0'][i]*0.025)+(df_urin['velo_lc0'][i-1])
    df_urin['velo_lc1'][i]=(df_urin['acc_lc1'][i]*0.025)+(df_urin['velo_lc1'][i-1])
    df_urin['velo_lc2'][i]=(df_urin['acc_lc2'][i]*0.025)+(df_urin['velo_lc2'][i-1])
    df_urin['velo_lc3'][i]=(df_urin['acc_lc3'][i]*0.025)+(df_urin['velo_lc3'][i-1])
    
df_def['velo_lc0'] = 0
df_def['velo_lc1'] = 0
df_def['velo_lc2'] = 0
df_def['velo_lc3'] = 0
for i in range (1,len(df_def)):
    df_def['velo_lc0'][i]=(df_def['acc_lc0'][i]*0.025)+(df_def['velo_lc0'][i-1])
    df_def['velo_lc1'][i]=(df_def['acc_lc1'][i]*0.025)+(df_def['velo_lc1'][i-1])
    df_def['velo_lc2'][i]=(df_def['acc_lc2'][i]*0.025)+(df_def['velo_lc2'][i-1])
    df_def['velo_lc3'][i]=(df_def['acc_lc3'][i]*0.025)+(df_def['velo_lc3'][i-1])
```

```python
X1 = df_ne.timestamp
X2 = df_urin.timestamp
X3 = df_def.timestamp

Y1 = df_ne.velo_lc0
Y2 = df_urin.velo_lc0
Y3 = df_def.velo_lc0

figure, axis = plt.subplots(3,1, figsize=(10,10))

axis[0].plot(X1,Y1)
axis[0].set_title("non extraction")

axis[1].plot(X2, Y2)
axis[1].set_title("urination")

axis[2].plot(X3, Y3)
axis[2].set_title("defecation")

plt.show()
```

## 5. Energy

```python
# e=1/2(m*v^2)
# m=w/9.81
#e = 0.5(w*v**2)
```

```python
df_ne['enr_lc0']= df_ne['velo_lc0'].apply(lambda x: 0.5*w_ne*(x**2))
df_ne['enr_lc1']= df_ne['velo_lc1'].apply(lambda x: 0.5*w_ne*(x**2))
df_ne['enr_lc2']= df_ne['velo_lc2'].apply(lambda x: 0.5*w_ne*(x**2))
df_ne['enr_lc3']= df_ne['velo_lc3'].apply(lambda x: 0.5*w_ne*(x**2))
# df_ne.head()

df_urin['enr_lc0']= df_urin['velo_lc0'].apply(lambda x: 0.5*w_urin*(x**2))
df_urin['enr_lc1']= df_urin['velo_lc1'].apply(lambda x: 0.5*w_urin*(x**2))
df_urin['enr_lc2']= df_urin['velo_lc2'].apply(lambda x: 0.5*w_urin*(x**2))
df_urin['enr_lc3']= df_urin['velo_lc3'].apply(lambda x: 0.5*w_urin*(x**2))
# df_urin.head()

df_def['enr_lc0']= df_def['velo_lc0'].apply(lambda x: 0.5*w_def*(x**2))
df_def['enr_lc1']= df_def['velo_lc1'].apply(lambda x: 0.5*w_def*(x**2))
df_def['enr_lc2']= df_def['velo_lc2'].apply(lambda x: 0.5*w_def*(x**2))
df_def['enr_lc3']= df_def['velo_lc3'].apply(lambda x: 0.5*w_def*(x**2))
# df_def.head()
```

```python
X1 = df_ne.timestamp
X2 = df_urin.timestamp
X3 = df_def.timestamp

Y1 = df_ne.enr_lc0
Y2 = df_urin.enr_lc0
Y3 = df_def.enr_lc0

figure, axis = plt.subplots(3,1, figsize=(10,10))

axis[0].plot(X1,Y1)
axis[0].set_title("non extraction")

axis[1].plot(X2, Y2)
axis[1].set_title("urination")

axis[2].plot(X3, Y3)
axis[2].set_title("defecation")

plt.show()
```

```python
X1 = df_ne.timestamp
X2 = df_urin.timestamp
X3 = df_def.timestamp

Y1 = df_ne.power_lc0
Y2 = df_urin.power_lc0
Y3 = df_def.power_lc0

figure, axis = plt.subplots(3,1, figsize=(10,10))

axis[0].plot(X1,Y1)
axis[0].set_title("non extraction")

axis[1].plot(X2, Y2)
axis[1].set_title("urination")

axis[2].plot(X3, Y3)
axis[2].set_title("defecation")

plt.show()
```
