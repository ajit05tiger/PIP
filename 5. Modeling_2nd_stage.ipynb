```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import stats
import os
import glob
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, roc_auc_score, plot_roc_curve
```

```python
df_main1 = pd.DataFrame(columns = ['name' ,'activity' ,'weight', 'duration','start_time' ,'mean_diff_sum', 'variance',  'maximim', 'median', 'skewness', 'kurtosis', '0.9_quantile', 'acceleration', 'velocity', 'energy', 'power', 'momentum'])
df_main1
```

```python
df_main=df_main1.copy()
```

```python
df_main=df_main[df_main['activity']!='non_elimination']
```

### Data cleaning

```python
df_main.shape
```

```python
df_main.dropna(inplace=True)
df_main.shape
```

```python
df_main=df_main[df_main['weight']>1000]
df_main.shape
```

```python
df_main=df_main[df_main['duration']<360]
df_main.shape
```

### Onehot encoding

```python
df_main=pd.get_dummies(df_main, columns=['activity'])
df_main.head(3)
```

```python
df_main.drop(['name', 'start_time', 'activity_urination'],axis=1,inplace=True)
df_main.head()
```

```python
df_shuffeled=df_main.sample(frac=1,random_state=101)
df_shuffeled.head(5)
```

### Data splitting to train, test and validation

```python
df_train=df_shuffeled[:int(len(df_shuffeled)*0.6)]
df_validate=df_shuffeled[int(len(df_shuffeled)*.6):int(len(df_shuffeled)*0.8)]
df_test=df_shuffeled[int(len(df_shuffeled)*0.8):]

print(df_train.shape)
print(df_validate.shape)
print(df_test.shape)
```

### Downsampling

```python
df_majority = df_train[df_train.activity_defecation==0]
df_minority = df_train[df_train.activity_defecation==1]
df_majority.shape,df_minority.shape
```

```python
from sklearn.utils import resample
df_downsampled = resample(df_majority,
                          replace=True,
                          n_samples=len(df_minority),
                          random_state=0)
```

```python
df_downsampled.shape
```

```python
df_train=pd.concat([df_downsampled, df_minority])
df_train
```

### Model preparation and result with all variables

```python
x_train = df_train.drop(columns='activity_defecation')
y_train = df_train['activity_defecation']

x_validate = df_validate.drop(columns='activity_defecation')
y_validate = df_validate['activity_defecation']

x_test = df_test.drop(columns='activity_defecation')
y_test = df_test['activity_defecation']
```

```python
def error_metrics(actual, predicted, x_data):
    print('Accuracy %age: ' + str(np.round(accuracy_score(actual, predicted)*100,2)))
    print('ROC-AUC Score: ' + str(np.round(roc_auc_score(actual, log_reg.predict_proba(x_data)[:,1])*100,2)))
    print('\nClassification Report: ')
    print(classification_report(actual, predicted))
    print('\nConfusion Matrix: ')
    print(confusion_matrix(actual, predicted))
    return 'Done'
```

### Training Data

```python
print('\nTraining\n--------------------------')

log_reg = LogisticRegression(random_state=10)
log_reg.fit(x_train, y_train)
y_train_pred=log_reg.predict(x_train)
error_metrics(y_train, y_train_pred, x_train)
print('\n ROC-AUC Curve for training set')
plot_roc_curve(log_reg, x_train, y_train)
plt.show()

print('\nDecision Tree\n----------------------')
classifier = DecisionTreeClassifier(criterion='entropy', random_state=10)
classifier.fit(x_train, y_train)
y_train_pred=classifier.predict(x_train)
error_metrics(y_train, y_train_pred, x_train)

print('\nRandom Forest\n-----------------------')
classifier = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=10)
classifier.fit(x_train, y_train)
y_train_pred=classifier.predict(x_train)
error_metrics(y_train, y_train_pred, x_train)
```

### Validation data

```python
print('\nValidation\n--------------------------')
print('\nLogistic Regression\n---------------')
log_reg = LogisticRegression(random_state=10)
log_reg.fit(x_train, y_train)
y_validate_pred=log_reg.predict(x_validate)
error_metrics(y_validate, y_validate_pred, x_validate)
print('\n ROC-AUC Curve for validate set')
plot_roc_curve(log_reg, x_validate, y_validate)
plt.show()

print('\nDecision Tree\n----------------------')
classifier = DecisionTreeClassifier(criterion='entropy', random_state=10)
classifier.fit(x_train, y_train)
y_validate_pred=classifier.predict(x_validate)
error_metrics(y_validate, y_validate_pred, x_validate)

print('\nRandom Forest\n-----------------------')
classifier = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=10)
classifier.fit(x_train, y_train)
y_validate_pred=classifier.predict(x_validate)
error_metrics(y_validate, y_validate_pred, x_validate)
```

### Test set

```python
print('\nTesting\n--------------------------')
print('\nLogistic Regression\n---------------')
log_reg = LogisticRegression(random_state=10)
log_reg.fit(x_train, y_train)
y_test_pred=log_reg.predict(x_test)
error_metrics(y_test, y_test_pred, x_test)
print('\n ROC-AUC Curve for test set')
plot_roc_curve(log_reg, x_test, y_test)
plt.show()

print('\nDecision Tree\n----------------------')
classifier = DecisionTreeClassifier(criterion='entropy', random_state=10)
classifier.fit(x_train, y_train)
y_test_pred=classifier.predict(x_test)
error_metrics(y_test, y_test_pred, x_test)

print('\nRandom Forest\n-----------------------')
classifier = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=10)
classifier.fit(x_train, y_train)
y_test_pred=classifier.predict(x_test)
error_metrics(y_test, y_test_pred, x_test)
```
