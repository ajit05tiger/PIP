```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import stats
import os
import glob
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, roc_auc_score, plot_roc_curve
```

### Creating main_file

```python
df_main1 = pd.DataFrame(columns = ['name' ,'activity' ,'weight', 'duration','start_time' ,'mean_diff_sum', 'variance',  'maximim', 'median', 'skewness', 'kurtosis', '0.9_quantile', 'acceleration', 'velocity', 'energy', 'power', 'momentum'])
df_main1
```

```python
def value_insertion(loc_csv):
    df=pd.read_csv(loc_csv)
    print(loc_csv)
    df_main1.loc[len(df_main1.index)] = [df['name'][0], df['activity'][0], df['weight'][0], df['timestamp'][len(df)-1], df['start_time'][0], df['diff_load_sum'].mean(), df['diff_load_sum'].var(), df['diff_load_sum'].max(), df['diff_load_sum'].median(), df['diff_load_sum'].skew(), df['diff_load_sum'].kurtosis(), df['diff_load_sum'].quantile(0.9), df['acc'].mean(), df['velo'].mean(), df['enr'].mean(), df['power'].mean(), df['mom'].mean()]
    return df_main1
```

```python
rootdir = "D:/Ajit/pip/sample_labeled5"
folder_list = []
for subdir, dirs, files in os.walk(rootdir):
    folder_list.append(subdir)
del folder_list[0]
```

### Converting 'urination' & 'defecation' into 'elimination'

```python
elemination_condition = ['urination', 'defecation']
df_main['activity1']=df_main['activity'].apply(lambda x: x if x not in elemination_condition else 'elemination')
df_main.head(3)
```

### Onehot encoding

```python
df_main=pd.get_dummies(df_main, columns=['activity1'])
df_main.head(3)
```

```python
df_main.drop(['name','activity', 'start_time', 'activity1_non_elimination'],axis=1,inplace=True)
df_main.head()
```

```python
df_shuffeled=df_main.sample(frac=1,random_state=101)
df_shuffeled.head(5)
```

### Data splitting to train, test and validation

```python
df_train=df_shuffeled[:int(len(df_shuffeled)*0.6)]
df_validate=df_shuffeled[int(len(df_shuffeled)*.6):int(len(df_shuffeled)*0.8)]
df_test=df_shuffeled[int(len(df_shuffeled)*0.8):]

print(df_train.shape)
print(df_validate.shape)
print(df_test.shape)
```

### Downsampling

```python
df_majority = df_train[df_train.activity1_elemination==1]
df_minority = df_train[df_train.activity1_elemination==0]
df_majority.shape,df_minority.shape
```

```python
from sklearn.utils import resample
df_downsampled = resample(df_majority,
                          replace=True,
                          n_samples=len(df_minority),
                          random_state=0)
```

```python
df_downsampled.shape
```

```python
df_train=pd.concat([df_downsampled, df_minority])
df_train
```
